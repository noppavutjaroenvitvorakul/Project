```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cicddos2019_dataset.csv')
```

```python
df = df.drop(df.columns[0], axis=1)
df
```

```python
# Correlation Matrix
corr = df.corr()
plt.figure(figsize=(100,100))
sns.heatmap(corr, annot=True, fmt='.2f')
plt.show()
```

```python
# Remove all features that have correlation value more than 0.9
df = df.drop(['Subflow Bwd Packets', 'Subflow Bwd Bytes'], axis=1)
```

```python
Y = df['Class']
X = df.drop(['Class'], axis=1)

# Select X only numerical data
X = X.select_dtypes(include=np.number)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, Y_train)
rf.score(X_test, Y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score, recall_score
y_pred = rf.predict(X_test)
cm = confusion_matrix(Y_test, y_pred)
print(cm)
accuracy_score(Y_test, y_pred)
```

```python
# Find precision and recall
from sklearn.metrics import  recall_score, precision_score
print("Recall = ",recall_score(Y_test, y_pred, average="binary", pos_label="Attack"))
print("Precision = ",precision_score(Y_test, y_pred, average="binary", pos_label="Attack"))
print("Accuracy = ",accuracy_score(Y_test, y_pred))
```
